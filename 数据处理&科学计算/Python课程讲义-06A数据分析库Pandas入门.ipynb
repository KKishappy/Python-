## 数据的合并

### 示例1

```python
import pandas as pd
df1 = pd.DataFrame({'lkey': ['AA', 'BB', 'UU','CC'],
                    'value': [1, 2, 3,5]})
df2 = pd.DataFrame({'rkey': ['AA', 'TT', 'BB','ZZ','CC'],
                    'value': [5, 6, 7,8,12]})

df = pd.merge(df1,df2,how='inner',left_on = 'lkey',right_on ='rkey') # 具有共有的 关键字 才保留
print('df:')
print(df)
```

```python
import pandas as pd
# 用于合并满足条件的代码数据
df1 = pd.DataFrame({'code': ['A01', 'A02', 'A03','A04']}) #,index=['PZ','PK','PQ','PA']
df2 = pd.DataFrame({'code': ['A01', 'A05', 'A02','A06','A04'],
                    'value': [5, 6, 7,8,12],
                    'height':[100,150,200,120,210]})

df = pd.merge(df1,df2,how='left',on = 'code') #left_on,right_on
print('df:')
print(df)
```
